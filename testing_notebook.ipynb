# Data class

```python
import pandas as pd


class Data:
    def __init__(self, data_file_name):
        self._xls = pd.ExcelFile(data_file_name) #validation
        self._dna_df_dict = self._xls.parse('DNAs', skiprows=18, index_col='DNA ID').to_dict('index')
        self._chemical_df_dict = self._xls.parse('Chemical', skiprows=18, index_col='Chemical ID').to_dict('index')
        self._signal_df_dict = self._xls.parse('Signal', skiprows=10, index_col='Signal ID').to_dict('index')
        self._vector_df_dict = self._xls.parse('Vector', skiprows=18, index_col='Vector ID').to_dict('index')
        self._supplement_df_dict = self._xls.parse('Supplement', skiprows=18, index_col='Supplement ID').to_dict('index')
        self._strain_df_dict = self._xls.parse('Strain', skiprows=18, index_col='Strain ID').to_dict('index')
        self._media_df_dict = self._xls.parse('Media', skiprows=10, index_col='Media ID').to_dict('index')
        self._measurement_df_dict = self._xls.parse('Measurement', skiprows=18, index_col='Measurement ID').to_dict('index')
        self._sample_design_df_dict = self._xls.parse('Sample Design', skiprows=18, index_col='Sample Design ID').to_dict('index')
        self._sample_df_dict = self._xls.parse('Sample', skiprows=18, index_col='Sample ID').to_dict('index')
        self._assay_df_dict = self._xls.parse('Assay', skiprows=18, index_col='Assay ID').to_dict('index')
        self._study_df_dict = self._xls.parse('Study', skiprows=18, index_col='Study ID').to_dict('index')

    @property
    def dna_df_dict(self):
        return self._dna_df_dict

    @dna_df_dict.setter
    def dna_df_dict(self, dna_dict):
        self._dna_df_dict = dna_dict

    @property
    def chemical_df_dict(self):
        return self._chemical_df_dict

    @chemical_df_dict.setter
    def chemical_df_dict(self, chemical_dict):
        self._chemical_df_dict = chemical_dict

    @property
    def signal_df_dict(self):
        return self._signal_df_dict

    @signal_df_dict.setter
    def signal_df_dict(self, signal_dict):
        self._signal_df_dict = signal_dict

    @property
    def vector_df_dict(self):
        return self._vector_df_dict

    @vector_df_dict.setter
    def vector_df_dict(self, vector_dict):
        self._vector_df_dict = vector_dict

    @property
    def supplement_df_dict(self):
        return self._supplement_df_dict

    @supplement_df_dict.setter
    def supplement_df_dict(self, supplement_dict):
        self._supplement_df_dict = supplement_dict

    @property
    def strain_df_dict(self):
        return self._strain_df_dict

    @strain_df_dict.setter
    def strain_df_dict(self, strain_dict):
        self._strain_df_dict = strain_dict

    @property
    def media_df_dict(self):
        return self._media_df_dict

    @media_df_dict.setter
    def media_df_dict(self, media_dict):
        self._media_df_dict = media_dict

    @property
    def measurement_df_dict(self):
        return self._measurement_df_dict

    @measurement_df_dict.setter
    def measurement_df_dict(self, measurement_dict):
        self._measurement_df_dict = measurement_dict

    @property
    def sample_design_df_dict(self):
        return self._sample_design_df_dict

    @sample_design_df_dict.setter
    def sample_design_df_dict(self, sample_design_dict):
        self._sample_design_df_dict = sample_design_dict

    @property
    def sample_df_dict(self):
        return self._sample_df_dict

    @sample_df_dict.setter
    def sample_df_dict(self, sample_dict):
        self._sample_df_dict = sample_dict

    @property
    def assay_df_dict(self):
        return self._assay_df_dict

    @assay_df_dict.setter
    def assay_df_dict(self, assay_dict):
        self._assay_df_dict = assay_dict

    @property
    def study_df_dict(self):
        return self._study_df_dict

    @study_df_dict.setter
    def study_df_dict(self, study_dict):
        self._study_df_dict = study_dict
```

## Assay

```python
def create_flapjack_assay(assay_key, flapjack_study_id):
    assay_name = Fj_data.assay_df_dict[assay_key]['Assay Name']
    assay_description = Fj_data.assay_df_dict[assay_key]['Description']
    assay_machine = Fj_data.assay_df_dict[assay_key]['Machine']
    assay_temperature = Fj_data.assay_df_dict[assay_key]['Temperature']
    return fj.create('assay', name=assay_name, 
        description=assay_description,
        machine = assay_machine,
        temperature = assay_temperature,
        study=flapjack_study_id)

#update hashmap adding supplement id
for assay_key in Fj_data.assay_df_dict:
    study_id = Fj_data.assay_df_dict[assay_key]['Study ID']
    flapjack_hash_study_id = hash_map[study_id]
    flapjack_assay_id = create_flapjack_assay(assay_key, flapjack_hash_study_id) #.id[0]
    hash_map[assay_key] = flapjack_assay_id.id[0]

print(hash_map)
```

## Sample

```python
def create_flapjack_sample(sample_key, flapjack_assay_id, flapjack_vector_id, flapjack_media_id, flapjack_strain_id, flapjack_supplement_id):
    #sample_design_id = Fj_data.sample_df_dict[sample_key]['Sample Design ID']
    row_number = Fj_data.sample_df_dict[sample_key]['Row']
    column_number = Fj_data.sample_df_dict[sample_key]['Column']

    return fj.create('sample',
        assay = flapjack_assay_id,
        row = row_number,
        col = column_number,
        vector = flapjack_vector_id,
        media = flapjack_media_id,
        strain = flapjack_strain_id,
        supplements = flapjack_supplement_id)

#update hashmap adding supplement id
for sample_key in Fj_data.sample_df_dict:
    assay_id = Fj_data.sample_df_dict[sample_key]['Assay ID']
    flapjack_hash_assay_id = hash_map[assay_id]
    media_id = Fj_data.sample_df_dict[sample_key]['Media ID']
    flapjack_hash_media_id = hash_map[media_id]
    strain_id = Fj_data.sample_df_dict[sample_key]['Strain ID']
    flapjack_hash_strain_id = hash_map[strain_id]
    vector_id = Fj_data.sample_df_dict[sample_key]['Vector ID']
    flapjack_hash_vector_id = hash_map[vector_id]
    sup_id = Fj_data.sample_df_dict[sample_key]['Supplement ID']
    flapjack_hash_sup_id = hash_map[sup_id]
    flapjack_sample = create_flapjack_sample(sample_key, 
                        flapjack_assay_id=flapjack_hash_assay_id, 
                        flapjack_vector_id=flapjack_hash_vector_id, 
                        flapjack_media_id=flapjack_hash_media_id,
                        flapjack_strain_id= flapjack_hash_strain_id,
                        flapjack_supplement_id=flapjack_hash_sup_id) #.id[0]
    hash_map[sample_key] = flapjack_sample.id[0]

print(hash_map)
```

```python
def create_flapjack_measurement(measurement_key, flapjack_sample_id, flapjack_signal_id):
    measurement_time = Fj_data.measurement_df_dict[measurement_key]['Time']
    measurement_value = Fj_data.measurement_df_dict[measurement_key]['Value']
    return fj.create('measurement', sample=flapjack_sample_id, 
                                    signal=flapjack_signal_id, 
                                    time=measurement_time, 
                                    value=measurement_value)

for measurement_key in Fj_data.measurement_df_dict:
    sample_id = Fj_data.measurement_df_dict[measurement_key]['Sample ID']
    flapjack_hash_sample_id = hash_map[sample_id]
    signal_id = Fj_data.measurement_df_dict[measurement_key]['Signal ID']
    flapjack_hash_signal_id = hash_map[signal_id]
    flapjack_measurement = create_flapjack_measurement(measurement_key, flapjack_hash_sample_id, flapjack_hash_signal_id)
    hash_map[key] = flapjack_measurement.id[0]

print(hash_map)
```
